# Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv("Israel-Palestine.csv")
```

```python
# Drop rows with null values
df= df.dropna(subset=['Active Military Personnel',
                                'Reserve Military Personnel',
                                'Literacy Rate (%)',
                                'IT Output (in USD)',
                                'Number of Tanks',
                                'Number of Submarines',
                                'Number of Armoured Vehicles'])
```

# Data Transformation 


```python
df.head()
```

```python
df['GDP (in USD)'] = df['GDP (in USD)'].replace({' billion': '', ',': ''}, regex=True).astype(float)
df.to_csv('updated_dataset.csv', index=False)
```

```python
df['Population'] = df['Population'].str.replace(',', '').astype(int)
df.to_csv('updated_dataset.csv', index=False)
```

```python
df['Active Military Personnel'] = df['Active Military Personnel'].str.replace(',', '').astype(int)
df.to_csv('updated_dataset.csv', index=False)
```

```python
df['Reserve Military Personnel'] = df['Reserve Military Personnel'].str.replace(',', '').astype(int)
df.to_csv('updated_dataset.csv', index=False)
```

```python
df['Number of Tanks'] = df['Number of Tanks'].str.replace(',', '').astype(int)
df.to_csv('updated_dataset.csv', index=False)
```

```python
df['Number of Armoured Vehicles'] = df['Number of Armoured Vehicles'].str.replace(',', '').astype(int)
df.to_csv('updated_dataset.csv', index=False)
```

```python
def convert_agricultural_output(value):
    if 'billion' in value:
        return float(value.replace(' billion', '').replace(',', '')) * 1e9
    elif 'million' in value:
        return float(value.replace(' million', '').replace(',', '')) * 1e6
    else:
        return float(value.replace(',', ''))

df['Agricultural Output (in USD)'] = df['Agricultural Output (in USD)'].apply(convert_agricultural_output)

df.to_csv('updated_dataset.csv', index=False)
```

```python
df['IT Output (in USD)'] = df['IT Output (in USD)'].replace({' billion': '', ',': ''}, regex=True).astype(float)
df.to_csv('updated_dataset.csv', index=False)
```
